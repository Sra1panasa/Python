```python
from transformers import AutoTokenizer
model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
input_ids=tokenizer("Hello, testing word indices ! Zwingmann")
print(input_ids)
#tokens after tokenizer
# 'z', '##wing', '##mann', these are additional tokens created by tokenizer
#['[CLS]', 'hello', ',', 'testing', 'word', 'indices', '!', 'z', '##wing', '##mann', '[SEP]']
# below input ids includes sub tokens also which tokenized by tokenizer.
```

```python
#Zwingmann word devided into 3 sub words- these 3 words should be treated together..to solve this problem of subtokens alignment we will convert
# input ids to word ids. 
#           ['[CLS]', 'hello', ',', 'testing', 'word', 'indices', '!', 'z', '##wing', '##mann', '[SEP]']
#input_ids- [101,      7592,  1010,   5604,     2773,   29299,    999, 1062,  9328,     5804,    102]
#word_ids-  [None,      0,     1,       2,        3,     4,        5,   6,    6,         6,     None]
tokens = tokenizer.convert_ids_to_tokens(input_ids["input_ids"])
print(tokens)
```

```python
# Using word_ids method to solve the problem of subtokens alignment
print(input_ids.word_ids())
```

**why we are converting to word_ids() from input ids:**

* after generating input ids by tokenizer there will be a increase in the tokens before tokenization and after tokenization.

* for example the word Zwingmann is splitted as 3 words.
* **tokens increase but labels remains same**.that is why we need to use word_ids which maps sub words of same word representing with same number.

* Example:

* [None, 0, 1, 2, 3, 4, 5, 6, 6, 6, None]

* Mageswaran explanation
*word ids represent the position of token before toenization
* after tokenization the number of tokens have increased while the labels for them remains same 
* we get the index of words and use it create a new list with respect to tokenised words

* WOON ---> woo, ##n
* index of woon is 1
* get label at index 1, say 'company'
* assign it to both woo and ##n
* (woo, company), (##n, company)
